```python
j=0
x1 = np.linspace(0,1,100)
cs1=CubicSpline(x1,np.zeros(100))
impressions=cs1(x1)
plt.plot(x1,cs1(x1))
print(impressions)        
```

```python
#normalizing x-axis in (0,1)range and adding all the curves for section 1
size=0
count=0
i=0
while i < len(df30.index)-1:
    y=[]
    x=[]
    if df30.loc[i,"section"]==1:
        j=0
        while df30.loc[i,"section"]==1 :
            if df30.loc[i,"Date"]!=df30.loc[i+1,"Date"]:
                break 
            else:
                print(i)
                y.append(df30.loc[i,"impressions"])
                i+=1
                j+=1
        x=np.linspace(0,1,j)
        print(x) 
        print(y)
        if len(x)<=1:
            continue
        cs = CubicSpline(x, y)  
        impressions=impressions+cs(x1)
        size=size+j
        count+=1
    i+=1   
        
```

# audio profile

```python
df_audio_profile=pd.read_csv("channel_2_29_30_22_24_show_2130_2200 (1).csv")
```

```python

df_audio_profile['date'] = df_audio_profile['date'].astype(str)
df_audio_profile['timestamp'] = df_audio_profile['timestamp'].astype(str)
df_audio_profile["date_time"]=df_audio_profile['date']+" " +df_audio_profile['timestamp']

df_audio_profile["timestamp"]=pd.to_datetime(df_audio_profile["timestamp"], format='%H:%M:%S')
df_audio_profile["timestamp"]= df_audio_profile["timestamp"].values.astype('<M8[m]')



df_audio_profile["date_time"]=pd.to_datetime(df_audio_profile["date_time"], format='%d-%m-%y %H:%M:%S')
df_audio_profile["date_time"]= df_audio_profile["date_time"].values.astype('<M8[m]')
```

```python
df_audio_profile["date"]=pd.to_datetime(df_audio_profile["date"], format='%d-%m-%y')
```

```python
df_audio_profile["Normal Speech"]=0
df_audio_profile["Silence/Quiet"]=0
df_audio_profile["Music/Special effects"]=0
df_audio_profile["Raised Voices"]=0
for index,rows in df_audio_profile.iterrows():
    if rows["audio_profile"]=="Normal Speech":
        df_audio_profile.set_value(index,'Normal Speech',1)
    elif rows["audio_profile"]=="Silence/Quiet":
        df_audio_profile.set_value(index,'Silence/Quiet',1)
    elif rows["audio_profile"]=="Music/Special effects":
        df_audio_profile.set_value(index,'Music/Special effects',1)
    elif rows["audio_profile"]=="Raised Voices":
        df_audio_profile.set_value(index,'Raised Voices',1)    
```

```python
#total 
df_audio_profile["total"]=1
```

```python
grouped_audio_profile = df_audio_profile.groupby(['Channel_id',"date","timestamp"])

max(grouped_audio_profile["Music/Special effects"].agg('count'))
```
